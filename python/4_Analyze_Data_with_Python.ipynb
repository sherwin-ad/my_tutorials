### What is Data Analysis?

**Data Analysis** is teh process of collecting , modelling, and analyzing data to exrtract insights that support decision-making.

**Statistics** is the thescience of collecting, organizing, analyzing, and interpreting data.

**Diagnostic analysis** is done to determine why something happened in the past.

**Descriptive analysis** looks at what has happened 

**Prescriptive analysis** wants to predict what is likely to happen in the future.


**Descriptive and diagnostic analysis** are common practices in business. Some companies do not have the manpower and resources to conduct a predictive analysis.

### Data Types

**1. Qualitative Data**
- Also called **categorical data**, qualitative data is a data type that isn't numerical.It represents characteristics. Though sometimes it can take on numerical values (for example, 1 for female and 0 for male), these numbers don’t have mathematical meaning.

**Types of Qualitative Data**

**Nominal**
- It is used to label variables that have no quantitative value. Mominal data has no order so if you would change the order of its values, the meaning would not change.

Examples: hair color, nationality

**Ordinal**
- It’s almost the same as nominal data, except that its order matters. Ordinal scales are usually used to measure non-numeric features like happiness, customer satisfaction, and so on, but the differences between the values is not really known.

Examples: educational background, satisfaction rating, soco-economic status

**2. Quantitative Data**
- Also known as **numerical data**, quantitative data is the type of data whose value is measured in the form of numbers or counts.

**Types of Quatitative Data**

**Discrete**
- It represents values that are distinct and separate. This type of data can’t be measured, but can be counted.

Examples: number of heads in 100 coin flips, number of people in a class, the number of fingers on your hands

**Continuous**
- It represents measurements whose values can’t be counted, but can be measured.

Examples: height, annual income

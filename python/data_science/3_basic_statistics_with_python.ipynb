- describes a sample

#### Common tools of Descriptive Statistics
##### Measures of central tendency
- tell us where the most values fall, or the typical value of a data 
set.

###### 1. Mode
- is the value that shows up te most number of times in a data set. In this case the mode is 5.
- is the most occuring number in a data set.

![image-20211116073203357](images/image-20211116073203357.png)

###### 2. Mean
- or the average, is the sum of all the numbers divided by the number of elements in the sample.
- is the average of all numbers in a data set.

![image-20211116074123420](images/image-20211116074123420.png)

###### 3. Median
- is the middle number of the data set.
- is the middle number in a sorted data set.

![image-20211116074256627](images/image-20211116074256627.png)

##### Measures of dispersion or spread

![image-20211116074729289](images/image-20211116074729289.png)

##### Common measures of dispersion
###### 1. Range
- is a difference between the high and low values.

![image-20211116074954900](images/image-20211116074954900.png)

###### 2. Variance
- measures how far a data set is spread out.

![image-20211118194510985](images/image-20211118194510985.png)

###### 3. Standard Deviation
- measures how far each observed value is from the mean.

![image-20211118194712751](images/image-20211118194712751.png)

![image-20211116075122416](images/image-20211116075122416.png)

**Note**
If the standard deviation is small, the data values are close to the mean value. If itis high, the data values are widely spread out from the mean value.

### Statistical Functions in Python

### Desctiptive Statistics

```python
import statistics
import array as arr

friends_usage = [5, 4, 3, 5, 8, 2, 5, 2, 5, 8, 10, 8, 7, 9, 10, 5, 7, 5, 7, 5, 4, 3, 5, 8, 4]

sample_data = arr.array("i",friends_usage)
dataset = sample_data[0:5]
sample_data_list = dataset.tolist()

mean = statistics.mean(friends_usage)
mode = statistics.mode(friends_usage)
median = statistics.median(friends_usage)
pstdev = round(statistics.pstdev(friends_usage),2)
pvariance = round(statistics.pvariance(friends_usage),2)
srange = (max(friends_usage)-min(friends_usage))

variance = round(statistics.variance(sample_data_list),2)
stdev = round(statistics.stdev(sample_data_list),2)

print("Descriptive Statistics")
print("Mean:", mean)
print("Mode:", mode)
print("Median:", median)
print("Range:", srange)
print("Population Variance:", pvariance)
print("Population Standard Deviation:", pstdev)
print("Variance:", variance)
print("Standard Deviation:", stdev)
```

### Displaying Statistical Data using Matplotlib

### Histogram
- is the most commonly used graph to show frequency distributions or how often each different value in a set of dtata occurs.

```python
import pandas as pd


f_usage = [5, 4, 3, 5, 8, 2, 5, 2, 5, 8, 10, 8, 7, 9, 10, 5, 7, 5, 7, 5, 4, 3, 5, 8, 4]

ser = pd.Series(f_usage)
sum_stat = ser.describe()

print(sum_stat)
```

```python
import matplotlib.pyplot as plt
import statistics as stat

fig = plt.figure(figsize=(16,9))

plt.hist(f_usage, bins=8, edgecolor="black")

plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.title('Friends Social Media Usage', fontsize=14)

plt.plot([stat.mean(f_usage), stat.mean(f_usage)], [0, 9], linestyle="--", color="r")
plt.plot([stat.median(f_usage), stat.median(f_usage)], [0, 10], linestyle="--", color="y")
plt.plot([stat.pstdev(f_usage), stat.pstdev(f_usage)], [0, 9], linestyle="--", color="g")
plt.plot([stat.mode(f_usage), stat.mode(f_usage)], [0, 9], linestyle="--", color="m")
plt.plot([stat.variance(f_usage), stat.variance(f_usage)], [0, 9], linestyle="--", color="c")

plt.text(8,8, 'mean', color='r')
plt.text(8,7.5, 'median', color='y')
plt.text(8,7, 'mode', color='m')
plt.text(8,6.5, 'stdev', color='g')
plt.text(8,6, 'variance', color='c')

plt.show()
```

### Horizontal Bar Graph with labels

```python
import matplotlib.pyplot as plt

country = sea['Country']

population = sea['Population (2020)']

fig, ax = plt.subplots(figsize =(16, 9))

ax.barh(country, population)

ax.set_title('2020 Population by Country (SEA)',loc ='center', )

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:ax.spines[s].set_visible(False)

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,str(round((i.get_width()), 2)),fontsize = 10, fontweight ='bold',color ='grey')

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding/spaces between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines
ax.grid(b = True, color ='grey',linestyle ='-.',linewidth = 0.5,
alpha = 0.2)

# Show top values
ax. invert_yaxis()

plt.show()
```
